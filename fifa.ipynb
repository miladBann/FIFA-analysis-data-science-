```python
df1 = pd.DataFrame(df, columns=['Name','Value', 'Wage'])
def string_to_integer(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        return float(x.replace('K', '')) * 1000 
    if 'M' in x:
        return float(x.replace('M', '')) * 1000000 
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000

Value = df1['Value'].replace('\€', '', regex=True).apply(string_to_integer)
Wage = df1['Wage'].replace('\€', '', regex=True).apply(string_to_integer)

df1['Value'] = Value
df1['Wage'] = Wage

df1['differense'] = df1['Value'] - df1['Wage']
df1.sort_values('differense', ascending=False)
df1
```
